```python
## importing nessesary Libraries
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
## Gradient Descent Alogorithm given data

## preprocess Data
data = pd.read_csv('testData-1.csv')
X = data.iloc[:,0]
Y = data.iloc[:,1]

# guesses for parameters
w_1 = 0.4  # Seed value for slope iteration
a = 0.1    # Learning rate 
e = 0.01   # Error value to break iteration prematurely

## construing the data model
def learning_model(X,Y,a, w_1,error):
    for i in range(0, 2000):
        sum = 0
        for j in range(0, len(X)):
            sum = sum + (w_1*X[j] - Y[j])*X[j]
        temp = w_1 - a*(sum/len(X))
        if abs(temp-w_1)< error:
            break
        w_1 = temp
    return (w_1) 
    
## saving the optimal slope for linear fit
s = learning_model(X,Y,a,w_1,e)
s_1 = s.round()
print("The optimized slope is {} which is approximately {}".format(s,s_1))

## creating function for linear fit
y = s*X

## plotting data w fit
plt.scatter(X, Y)  # plot data
plt.plot(X, y, '--r') # dashed red

## labeling plot
plt.xlabel("Indicies of data points")
plt.ylabel("Data values")
plt.title("Univariate Fit")
```
